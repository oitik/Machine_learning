## Two option for subplots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))
ax1.plot(x, x**2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
ax[0, 0].plot(x, x**2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(butter_prices.keys(), butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Which one should you use? (pyplot vs matplotlib OO method?)
* when plotting something quickly its okay to use pyplot mehtod
* when plotting something more advanced, use the OO method


```python
hrt_dis.head(4
            )
```

```python
# over 50 ages data
over_50 = hrt_dis[hrt_dis['age'] > 50 ]
over_50.head(5)
```

```python
over_50.size, hrt_dis.size, len(over_50), len(hrt_dis)
```

```python
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target');
# pyplot method
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (7, 5))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax,
             ylabel = 'cholestorol');
# ax.set(xlim = [45, 85])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'],
                     c = over_50['target']);
# customize the plot
ax.set(xlabel = 'Age', ylabel = 'Cholesterol',
       title = 'Heart Disease and Cholesterol Levels',
       );
# add a legend
ax.legend(*scatter.legend_elements(), title = 'target');

# add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle = '--');
```

```python
over_50.head()
```

```python
# subplot option 1
fig, ((ax1, ax2)) = plt.subplots(ncols = 2,
                                 figsize = (12, 5))
ax1.scatter(over_50['age'], over_50['chol'],
           c = over_50['target']);
ax1.set(xlabel = 'Age', ylabel = 'Cholesterol',
       title = 'Heart Disease and Cholesterol Levels',
       );
ax1.legend(*scatter.legend_elements(), title = 'target');
ax1.axhline(over_50['chol'].mean(),
           linestyle = '--');


# axis 2
ax2.scatter(over_50['age'], over_50['thalach'],
           c = over_50['target']);
ax2.set(xlabel = 'Age',
        ylabel = 'Max Heart Rate',
        title = 'Heart Disease and Max Heart Rate',
        );
ax2.legend(*scatter.legend_elements(), title = 'target');
ax2.axhline(over_50['thalach'].mean(),
           linestyle = '--');

fig.suptitle('Heart Disease Anlysis', fontsize = 16, fontweight = 'bold', c = 'blue');
```

```python
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'],
                     c = over_50['target'],
                     cmap = 'winter');
# customize the plot
ax.set(xlabel = 'Age', ylabel = 'Cholesterol',
       title = 'Heart Disease and Cholesterol Levels',
       );
# add a legend
ax.legend(*scatter.legend_elements(), title = 'target');

# add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle = '--');
```

```python
# subplot option 1
fig, (ax1, ax2) = plt.subplots(nrows= 2,
                               figsize = (10, 10))

ax1.scatter(over_50['age'],
            over_50['chol'],
            c = over_50['target'],
            cmap = 'winter');

ax1.set(xlabel = 'Age', ylabel = 'Cholesterol',
       title = 'Heart Disease and Cholesterol Levels',
       xlim = [50, 80]);



ax1.axhline(over_50['chol'].mean(),
           linestyle = '--');

ax1.legend(*scatter.legend_elements(prop = 'colors'), title = 'Target')
# leg.legendHandles[0].set_color('red')
# leg.legendHandles[1].set_color('yellow'

# axis 2
ax2.scatter(over_50['age'], over_50['thalach'],
           c = over_50['target'],
           cmap = 'winter');
ax2.set(xlabel = 'Age',
        ylabel = 'Max Heart Rate',
        title = 'Heart Disease and Max Heart Rate',
        xlim = [50, 80],
        ylim = [60, 200]
        );
ax2.legend(*scatter.legend_elements(), title = 'target');
ax2.axhline(over_50['thalach'].mean(),
           linestyle = '--');

fig.suptitle('Heart Disease Anlysis', fontsize = 16, fontweight = 'bold', c = 'blue');
```
