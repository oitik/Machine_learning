```python
x = [1, 2, 3, 4]
y = [20, 14, 17, 18]
```

## Class Test Marks plot

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols = 2,
                                 nrows = 2,
                                 figsize = (12, 10))
# data set for EEE ct
ax1.plot(x, y, linestyle = '-', linewidth = 2.5, color = 'red');

ax1.set(title = 'EEE Class Test Marks',
        xlabel = 'ct_number',
        ylabel = 'ct_marks',
        ylim = [0, 20]);
ax1.legend(['Marks', 'Average']);
# draw the average line
ax1.axhline(15, linestyle = '--');


# data set for Algorithm ct
ax2.plot(x, [13, 19, 12, 12], linestyle = '-', linewidth = 2.5, color = 'blue');

ax2.set(title = 'Algorithm Class Test Marks',
        xlabel = 'ct_number',
        ylabel = 'ct_marks',
        ylim = [0, 20]);
ax2.legend(['Marks', 'Average']);
# draw the average line
ax2.axhline(15, linestyle = '--');


# data set for Automata ct
ax3.plot(x, [19, 17, 16, 18], linestyle = '-', linewidth = 2.5, color = 'green');

ax3.set(title = 'Automata Class Test Marks',
        xlabel = 'ct_number',
        ylabel = 'ct_marks',
        ylim = [0, 20]);
ax3.legend(['Marks', 'Average']);
# draw the average line
ax3.axhline(15, linestyle = '--');

# data set for Digital techniques ct
ax4.plot(x, [18, 17, 11, 16], linestyle = '-', linewidth = 2.5, color = 'violet');

ax4.set(title = 'Digital Techniques Class Test Marks',
        xlabel = 'ct_number',
        ylabel = 'ct_marks',
        ylim = [0, 20]);
ax4.legend(['Marks', 'Average']);
# draw the average line
ax4.axhline(15, linestyle = '--');

fig.suptitle('Class Test Marks Graph', fontsize = 20, fontstyle = 'italic', color = 'black');
```

```python
over_60 = hrt_dis[hrt_dis['age'] >= 60]
len(over_60)
```

```python
over_60.head()
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-dark-palette')
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_60['age']
                    ,y = over_60['trestbps']
                    ,c = over_60['target']
                    ,cmap = 'magma');
# customize the plot
ax.set(xlabel = 'Age', ylabel = 'Blood Pressure');
ax.set_title('Heart Disease and Blood Pressure', color = 'red')
# add a legend
ax.legend(*scatter.legend_elements(), title = 'target');

# add a horizontal line
ax.axhline(over_60['trestbps'].mean(),
           linestyle = '--');
```
